```python
import random 
MAX_LINES=3
MAX_BET=100
MIN_BET=1

ROWS=3
COLS=3

symbol_count={
    "A": 2,
    "B": 4,
    "C": 6,
    "D": 8
}
symbol_value={
    "A":5,
    "B":4,
    "C":3,
    "D":2,
}
def check_winnings(columns,lines,bet,values):
    winnings=0
    winning_lines=[]
    for line in range(lines):
        symbols=columns[0][line];
        for column in columns:
            symbol_to_check=column[line]
            if symbols!=symbol_to_check:
                break
        else:
            winnings+=values[symbols]*bet
            winning_lines.append(line+1)
            
    return winnings,winning_lines
            
            
    
def get_spin(rows,cols,symbols):
    all_symbols=[]
    for symbol,symbol_count in symbols.items():
        for _ in range(symbol_count):
            all_symbols.append(symbol)
    columns=[]
    for _ in range(cols):
        column=[]
        current_symbols= all_symbols[:] #slice operator, it is used to copy the list and they become independt of eachother
        for _ in range(rows):
            value=random.choice(current_symbols)
            current_symbols.remove(value)
            column.append(value)
        columns.append(column)
    return columns
            
def printslot(columns):
    #transpose the matrxi
    for row in range(len(columns[0])):
        #we are doing all this to print the numbers like A|B|C to get two bars
        #enumerate will print the index as well as the value
        for i,column in enumerate(columns):
            if i!=len(columns)-1:
                print(column[row],end=" | ")
            else:
                print(column[row],end=" ")
        print()
            
    

def deposit():
    while True:
        amount=input("What would you like to deposit? $");
        #check if the amount is number
        if amount.isdigit():
            amount=int(amount)
            if amount>0:
                break
            else:
                print("amount must be greater than 0")
        else:
            print("please enter a number")
    return amount
#to call a function just write its name 
def get_num_of_lines():
    while True:
        #adding the int MAX_LINES to string by  "+str(the int)+"
        lines=input("enter number of lines to bet on (1-" + str(MAX_LINES)+") ? ")
        #check if the amount is number
        if lines.isdigit():
            lines=int(lines)
            if 1<=lines<=MAX_LINES:
                break
            else:
                print("enter valid number of lines")
        else:
            print("please enter a number")
    return lines

def get_bet():
    while True:
        amount=input("What would you like to bet on each line? $");
        #check if the amount is number
        if amount.isdigit():
            amount=int(amount)
            if MIN_BET<=amount<=MAX_BET:
                break
            else:
                #another way of adding Varibel to the strings, automatically gets converted to string
                print(f"amount must be between ${MIN_BET}-${MAX_BET} ")
        else:
            print("please enter a number")
    return amount

def spin(balance):
    lines=get_num_of_lines()
    #checking if the bet is within the deposit
    while True:
        bet=get_bet();
        total_bet=bet*lines
        if total_bet>balance:
            print(f"you dont have enough to bet that amount, your current balance is: ${balance}" )
        else:
            break
            
    print(f"you are betting ${bet} on {lines} lines. total bet is equal to: ${bet*lines}")
    solts=get_spin(ROWS,COLS,symbol_count)
    printslot(solts)
    winnings,winning_lines=check_winnings(solts,lines,bet,symbol_value);
    print(f"you won ${winnings}")
    print(f"you won on : ", *winning_lines)
    return winnings-total_bet
    
def main():
    balance=deposit()
    while True:
        print(f"current balance is: ${balance}")
        ans=input("press enter to play (q to quit).")
        if ans=="q":
            break
        balance+=spin(balance)
    print(f"you left with {balance}")
    
main()
```
